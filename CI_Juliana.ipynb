# 1) When is the best time of the year to book rooms?

## Based on demand - the time of the year the hotel is less busy
Higher availability = lower occupancy

```python
### Checking if arrival_date_month has the correct values and data type
df.arrival_date_month.unique()
```

```python
### Converting arrival_date_month data type to categorical for ordering
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

df['arrival_date_month'] = pd.Categorical(df['arrival_date_month'],
                                          categories = month_order,
                                          ordered=True)
```

```python
import matplotlib.pyplot as plt

### Grouping data by month
monthly_counts = df.groupby('arrival_date_month').size()

### Plotting the bar chart
monthly_counts.plot(kind='bar', color='purple', figsize=(10,6))
plt.title('Number of Bookings by Arrival Month')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle=':', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
### Creating a new arrival_month column

# Creating a dict with the correct order of the months
month_map= {'January':1, 'February':2, 'March':3, 'April':4, 'May': 5, 'June':6,
            'July':7, 'August':8, 'September':9, 'October':10, 'November':11, 'December':12}

# Creating a new column associating the numbers with the months
df['arrival_month_num'] = df['arrival_date_month'].map(month_map)

# Creating a new column for month-year, ignoring the day
df['arrival_month'] = pd.to_datetime(dict(
    year=df['arrival_date_year'],
    month=df['arrival_month_num'],
    day=1 #Ignoring the day
))
```

```python
# Grouping by monthly arrival
monthly_counts = df.groupby(df['arrival_month']).size()

# Creating the bar plot
fig, ax = plt.subplots(figsize=(14, 6))
monthly_counts.plot(kind='bar', ax=ax, color='purple')

# Formatting x-axis labels to show "Jan 2019", "Feb 2019", etc.
formatted_labels = monthly_counts.index.strftime('%b %Y')  # %b = short month name, %Y = full year
ax.set_xticklabels(formatted_labels, rotation=45, ha='right')

# Setting labels and title
plt.title('Monthly Bookings Over Time')
plt.xlabel('Arrival Date (Month-Year)')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
### Separating the dataset according to the hotel

# Grouping data by hotel type and arrival month
grouped = df.groupby(['arrival_month', 'hotel']).size().unstack(fill_value=0)

# Plotting grouped bar chart
fig, ax = plt.subplots(figsize=(14, 6))

# Creating bar positions
bar_width = 10
dates = grouped.index

# Plotting each hotel type
ax.bar(dates - pd.Timedelta(days=bar_width/2), grouped['City Hotel'], width=bar_width, label='City Hotel', color='lightblue')
ax.bar(dates + pd.Timedelta(days=bar_width/2), grouped['Resort Hotel'], width=bar_width, label='Resort Hotel', color='lightgreen')

# Formatting
ax.set_title('Monthly Bookings by Hotel Type')
ax.set_xlabel('Arrival Month')
ax.set_ylabel('Number of Bookings')
ax.set_xticks(dates)
ax.set_xticklabels(dates.strftime('%b %Y'), rotation=45, ha='right')
ax.grid(axis='y', linestyle='--', alpha=0.5)
ax.legend()

plt.tight_layout()
plt.show()
```

# 5) Tell me more about my top agent during this time

```python
# Filtering for only rows where agent is not null
df_agent_filtered = df[df['agent'].notnull()]

# Comparing the number of nulls in the agent column before and after filtering
print(df['agent'].isnull().sum())

print(df_agent_filtered['agent'].isnull().sum())
```

```python
agent_counts = df_agent_filtered['agent'].value_counts()
print(agent_counts)
```

Travel agency 9 is the most successful in making bookings, followed by agency 240 and 1.

# 6) Model to classify the likelihood of cancellation at the hotels

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df.info()
```

#### Defining the features and target

Picking the columns that will be my input variables and the target (cancellations).

```python
# Define features and target

features = ['lead_time', 'adr', 'stays_in_week_nights', 'stays_in_weekend_nights', 'total_of_special_requests', 'hotel']

# dropping all missing rows, ensuring consistent number in columns used
df2_clean = df.dropna(subset=features + ['is_canceled'])

X = df2_clean[features]
y = df2_clean['is_canceled']
```

#### Splitting the dataset

```python
# Separating the data into training and test data, separating by hotel (categorical), to ensureboth sets have a similar distribution of hotels

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=df2_clean['hotel'], test_size=0.2, random_state=42 #fixing the split to get the same every time the code is run
)
```

```python
# Verifying consistency of shapes, ensuring number of samples in the features and target are always aligned
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

#### Preprocess features

```python
# Separating categorical and numerical features
categorical_features = ['hotel']
numerical_features = ['lead_time', 'adr', 'stays_in_week_nights', 'stays_in_weekend_nights', 'total_of_special_requests']

# Column transformer for scaling numeric and encoding categorical
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(drop='first'), categorical_features)
])

#Preprocessor step takes raw input data and performs imputation (filling missing values), scaling (normalising numeric features), and encoding (converting categorical to numeric)
#It outputs clean, numeric feature matrix ready for modeling.
```

#### Creating a pipeline with logistic regression

```python
model = Pipeline(steps=[
    ('preprocessor', preprocessor), # Applies all preprocessing (imputation, scaling, encoding)
    ('classifier', LogisticRegression(max_iter=1000, random_state=42)) # Receives the preprocessed numeric features and fits the logistic regression model on training data
])
```

#### Training a model

```python
model.fit(X_train, y_train)
```

#### Evaluating the model

```python
y_pred = model.predict(X_test)

print("Classification Report:\n")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:\n")
print(confusion_matrix(y_test, y_pred))
```
